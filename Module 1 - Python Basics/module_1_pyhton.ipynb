Exercise: Calculate compound interest for principal=1000, rate=5%, time=3 years.

```python
# Compound interest caculation
principal = 1000 # initial amount
rate = 5 #annual interest rate in percentage 
time = 3 # time in years 

amount = principal * (1+ rate/100)** time
compound_interest = amount - principal

print("Compound Interest:", f"{compound_interest:.2f}")
```

Exercise: Create a dictionary for 3 books with title, author, price. Print all titles.

```python
# dictionary of 3 books with title, author, price
books = {
    "book1": {"title": "Python Programming", "author": "John Doe", "price": 29.99},
    "book2": {"title": "Data Science Basics", "author": "Jane Smith", "price": 39.99},
    "book3": {"title": "Machine Learning", "author": "Alice Johnson", "price": 49.99}
}
# print all th titles of the books

for book in books.values():
    print("Book Title:", book["title"])

for book_id, book_info in books.items():
    print(f"{book_id}: {book_info['title']} by {book_info['author']} - ${book_info['price']}")
```

Exercise: Write a loop to print only even numbers between 1 and 20.

```python
# Write a loop to print only even numbers between 1 and 20.

for i in range (1, 21):
    if (i%2 == 0):
        print(i)
```

```python
# using break continue and pass

for i in range(1, 11):
    if i == 7:
        break  # exit the loop when i is 5
    if i % 2 == 0:
        continue  # skip even numbers
    print(i)
print("for pass:")

# for pass
for i in range(1, 6):
    if i == 3:
        pass  # placeholder for future code
    print(i)
# what does pass do in python
# The pass statement is a null operation in Python. It does nothing when executed.
# It is used as a placeholder where syntactically a statement is required but no action is desired.

# In the above example, when i equals 3, the pass statement is executed.
# However, since pass does nothing, the loop continues to print all numbers from 1 to 5 without any interruption.
```

Lesson 1.6: Functions and Lambda

Concepts

Functions: reusable code

Parameters, return values

Lambda: anonymous functions


```python
#function 
def square(x):
    return x**2

print("Square of 5", square(5))

#Lambda Function
double = lambda x:x*2
print("Double 10:", double(10))

#explain the difference between a regular function and a lambda function in Python
# A regular function in Python is defined using the def keyword and can contain multiple statements,
# whereas a lambda function is an anonymous, single-expression function defined using the lambda keyword.
# Regular functions can have multiple lines of code and can include complex logic,
# while lambda functions are limited to a single expression and are typically used for short, simple operations.
# Regular functions have a name and can be called multiple times,
# whereas lambda functions are often used as throwaway functions and are not bound to a name unless assigned to a variable.
```

Exercise: Write a function that takes a number and returns "Even" or "Odd".

```python
def even_or_odd(num):
    if num% 2 == 0:
        return "Even"
    else:
        return "Odd"
    
print("Even or Odd 7:", even_or_odd(7))
```

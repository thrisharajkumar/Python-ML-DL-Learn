Exercise 1: Find mean, median, mode, variance, and standard deviation for the dataset [10, 20, 20, 30, 40, 50, 50, 50].

```python
import statistics as stats 
import numpy as np 

data = [10,20, 20, 30, 40, 50, 50 , 50]

print("Mean:", stats.mean(data))
print("Median:", stats.median(data))
print("Mode:", stats.mode(data))
print("Variance:", stats.variance(data))
print("Standard Deviation:", stats.stdev(data))

# Using NumPy
print("Mean (NumPy):", np.mean(data))
print("Std (NumPy):", np.std(data))
```

Exercise 4: Simulate rolling a dice 1000 times and find probability of getting a 6.

```python
import numpy as np

rolls = np.random.randint(1, 7, 1000)
prob_six = np.sum(rolls == 6) /1000
print("Probablity of 6", prob_six)
```

Exercise 5: Find probability of rolling an even number.

```python
prob_even = np.sum(rolls % 2 ==0)/1000
print ("Probability of even:", prob_even)
```

exercise 5: Find probability of rolling an even number.

```python
flitered = rolls[rolls > 4]
prob_cond = np.sum(flitered == 6) / len(flitered)
print("Conditional Probability of 6 given >4:", prob_cond)
```

Exercise 11: Create DataFrame with missing values and clean it.

```python
data = {"Name": ["Alice", "Bob", "Charlie", "David"],
         "Age":[25,np.nan,23,22],
        "Score":[85,90,np.nan,95]}
df = pd.DataFrame(data)
print("Original DF:\n", df)

# Fill missing values with mean 
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Score'].fillna(df['Score'].median(), inplace=True)
print("Cleaned DF:\n", df)
```

Part 6: Outliers Detection

Exercise 13: Detect outliers using IQR.

```python
data = pd.DataFrame({"Score":[50,55,60,65,70,75,200]})
Q1 = data['Score'].quantile(0.25)
Q3 = data['Score'].quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data['Score'] < Q1 - 1.5*IQR) | (data['Score'] > Q3 + 1.5*IQR)]
print("Outliers:\n", outliers)
```

Exercise 14: Replace outliers with median.

```python
median = data['Score'].median()
data['Score'] = np.where((data['Score'] < Q1 - 1.5*IQR) | (data['Score'] > Q3 + 1.5*IQR),
                         median, data['Score'])
print("After replacing outliers:\n", data)
```

Part 7: Exploratory Data Analysis (EDA) Mini Project

Goal: Perform EDA on a dataset of 50 students with Name, Age, Score, Gender.

Steps:

Generate random dataset

Check data types and missing values

Summary statistics

Visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

#generate dataset

names = [f"Student{i}" for i in range(1,51)]
ages = np.random.randint(18,26,50)
scores = np.random.randint(0,101,50)
genders = np.random.choice(["M","F"],50)
df = pd.DataFrame({"Name":names, "Age":ages, "Score":scores, "Gender":genders})
print(df.head())

# Check data types and missing values
print(df.info())
print(df.isnull().sum())

# Summary stats
print(df.describe())

# Visualizations
plt.hist(df['Score'], bins=10)
plt.title("Score Distribution")
plt.show()

sns.boxplot(x='Gender', y='Score', data=df)
plt.title("Score by Gender")
plt.show()

sns.scatterplot(x='Age', y='Score', hue='Gender', data=df)
plt.title("Score vs Age")
plt.show()

# Correlation
print("Correlation:\n", df[['Age','Score']].corr())
```
